```python
import pandas as pd
import matplotlib.pyplot as plt

#data was downloaded from Kaggle : https://www.kaggle.com/datasets/lakshmi25npathi/online-retail-dataset/
# Read the data into a DataFrame
data = pd.read_csv('online-retail.csv')

# Group the data by product and calculate the total number of sales for each product
product_sales = data.groupby('Description').agg({'Quantity': 'sum'})

# Sort the products by total sales in descending order
product_sales = product_sales.sort_values(by='Quantity', ascending=False)

# Select the top 10 products
top_10_products = product_sales.head(10)

# Create a bar chart of the top 10 products
top_10_products.plot(kind='bar', y='Quantity', legend=False)
plt.ylabel('Total Sales')
plt.xlabel('Product')
plt.title('Top 10 Most Popular Products')
plt.show()
```
